```python
def adiciona_valor_logico(objeto):
    if ('show' in objeto.lower() and 'musica' in objeto.lower()) or 'banda musical' in objeto.lower() or 'cantor' in objeto.lower() or ('banda' in objeto.lower() and 'larga' not in objeto.lower() and 'bandarra' not in objeto.lower() and 'aquisição' not in objeto.lower()):
        return True
    else:
        return False
```

```python
indexes = []
for index, linha in enumerate(dados_novos['pop']):
    if '(' in str(linha):
        print(index)
        indexes.append(index)

from regex import search
def extrair_pop(valor):
    if type(valor) == str:
        match = search(r'(\d+\.\d+|\d+)', valor)
        if match:
            return int(match.group())
        else:
            return None
    else:
        pass

dados_novos.loc[dados_novos.iloc[indexes].index.unique(), 'pop'] = dados_novos.loc[dados_novos.iloc[indexes].index.unique(), 'pop'].str.replace('.','').apply(extrair_pop)


dados_novos['pop'] = dados_novos['pop'].fillna(0).astype(int)

dados_novos = dados_novos.loc[dados_novos['pop'] != 0]
```
